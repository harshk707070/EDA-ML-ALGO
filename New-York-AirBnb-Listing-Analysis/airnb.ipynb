#### 5. Data Analysis

#### Univariate Analysis - Single variable analysis

```python
# identify outliers in price
df = data[data['price'] < 1500]
sns.boxplot(data=df, x='price')
```

```python
# Analyse price distribution
plt.figure(figsize=(10, 6))
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
sns.histplot(data=df, x='price', bins=100)
plt.show()
```

```python
# Price distribution
plt.figure(figsize=(10, 6))
plt.title('Availability_365 Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
sns.histplot(data=df, x='availability_365')
plt.show()
```

```python
data.dtypes
```

```python
# Average price per bed in each neighbourhood group
df.groupby('neighbourhood_group')['price_per_bed'].mean()
```

```python
df.groupby('neighbourhood_group')['price'].mean()
```

Feature Engineering

```python
# price_per_bed

df['price_per_bed'] = df['price']/df['beds']
df.head()
```

#### Bivariate analysis

how one variable depends on another variable

```python
df.columns
```

```python
# price dependency on neighbourhood group and room type
sns.barplot(data=df, x='neighbourhood_group', y='price', hue = 'room_type')
```

```python
# number of reviews and price
plt.figure(figsize=(8, 6))
plt.title('Number of Reviews vs Locality')
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

```python
df.dtypes
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
```

```python
# Geographical distribution of listings
plt.figure(figsize=(8,6))
plt.title('Geographical Distribution of Listings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.show()
```

```python
df.dtypes
```

```python
# Heat map of correlations
# correlation of one column with other columns (Numericals only)

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 
           'number_of_reviews', 'reviews_per_month', 
           'availability_365', 'beds']]

# Computing correlation matrix
corr_matrix = corr.corr()

plt.figure(figsize=(10, 6))
plt.title('Correlation Heatmap', fontsize=14)
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.show()
```
